# File handling write and append mode

```python
with open("student.txt", "w") as f:
    f.write("Hello World")
```

```python
with open("student.txt", "a") as f:
    f.write("\nI am learning python")
```

```python
with open("student.txt", "a") as f:
    f.writelines(["\nPython File Handling\n","SPRK Tech"])
```
