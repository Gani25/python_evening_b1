```python
student_info = {} # dict
with open("student_marks.csv") as student_data:
    for row in student_data:
        student_name, subject,mark = row.split(",")
        
        print("-----------------------")
        print("Name: ",student_name)
        print("Subject: ",subject)
        print("Marks: ",mark)
        print("-----------------------")
```

```python
student_info = {} # dict
with open("student_marks.csv") as student_data:
    for row in student_data:
        student_name,subject,mark = row.split(",")
        mark = int(mark)
        if student_name in student_info:
            student_info[student_name].append(mark)
        else:
            student_info[student_name] = list([mark])
        
print(student_info)
```

```python
for k,v in student_info.items():
    print("---------------------------")
    print(f"Student Name = {k}")
    print(f"Marks = {v}")
    print(f"Sum Of Marks = {sum(v)}")
    print(f"Minimum Of Marks = {min(v)}")
    print(f"Maximum Of Marks = {max(v)}")
    print(f"Average Of Marks = {round(sum(v)/len(v),2)}")
```
