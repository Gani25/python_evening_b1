# Find Average. Min, And Max Rating Of All Movies

```python
print("Max IMDB Rating of all movies =",df_movies["imdb_rating"].max(),"star")
print("Min IMDB Rating of all movies =",df_movies["imdb_rating"].min(),"star")
print("Average IMDB Rating of all movies =",df_movies["imdb_rating"].mean().round(2),"star")
```
