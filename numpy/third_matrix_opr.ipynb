```python
import numpy as np
```

```python
q1_sales = np.array([
    [10,15,6],
    [25,12,18],
    [28,10,5]
])

q1_sales
```

```python
q2_sales = np.array([
    [35,12,8],
    [30,15,14],
    [10,0,5]
])

q2_sales
```

# Total unit sold in both quarters

```python
q1_sales + q2_sales
```

# Sales unit growth from q1 to q2

```python
q2_sales-q1_sales
```

# Percentage growth in sales from q1 to q2

```python
(q2_sales-q1_sales)/q1_sales * 100
```

# Find Quarter 1 Revenue

```python
prices = np.array([
    [110,105,98],
    [50,48,51],
    [10,8,12]
])

prices
```

```python
q1_revenue = prices * q1_sales
q1_revenue
```

```python
q1_discount = q1_revenue * 0.2
q1_discount
```

```python
q1_net_revenue = q1_revenue - q1_discount
```

```python
q1_net_revenue
```

```python
q1_discount.sum()
```

```python
q1_net_revenue.sum()
```

# Stacking

```python
student = np.array(
[
    [101,"Rohit",25],
    [102,"SHivam",28],
    [103,"Pranjal",29],
    [104,"Rohini",18],
]

)

student_info = np.array([
    [101,"Male",85.5],
    [102,"Male",35.55],
    [104,"Female",98.99],
    [103,"Female",38.99],
])

print(student)
print(student_info)
```

```python
np.hstack((student,student_info))
```

```python
more_student = np.array([[105,"Ashish",30],[110,"Shubham",45]])
more_student
```

```python
np.vstack((student,more_student))
```
