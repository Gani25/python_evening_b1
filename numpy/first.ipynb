## Numpy as Space Efficient

```python
import sys
```

```python
numbers = list(range(1,10001))

print(f"First 5 elements = {numbers[:5]}")
print(f"Last 5 elements = {numbers[-1:-6:-1]}")
print(f"Individual Size of element {sys.getsizeof(numbers[0])} bytes")
print(f"Total Size of python list {sys.getsizeof(numbers[0]) * len(numbers)} bytes")
```

```python
arrays = np.arange(1,10001, dtype="int64")

print(f"First 5 elements of numpy arrays = {arrays[:5]}")
print(f"Last 5 elements of numpy arrays= {arrays[-1:-6:-1]}")
print(f"Individual Size of element in numpy arrays {arrays.itemsize} bytes")
print(f"Total size of numpy arrays {arrays.nbytes} bytes")
```

```python
l1 = list(range(1,100000001))
l2 = list(range(1,100000001))

print(f"First 5 element of l1 = {l1[:5]}")
print(f"Last 5 element of l1 = {l1[-1:-6:-1]}")
print(f"First 5 element of l2 = {l2[:5]}")
print(f"Last 5 element of l2 = {l2[-1:-6:-1]}")
```

```python
list3 = []
start = time.time()
ind = 0
while ind < len(l1):
    list3.append(l1[ind]+l2[ind])
    ind = ind + 1 

end = time.time()

print(f"First 5 element of list3 = {list3[:5]}")
print(f"Last 5 element of list3 = {list3[-1:-6:-1]}")
print(f"Time Take to complete operation of 2 list addition is {end-start} sec")
```

```python
arr1 = np.arange(1,100000001)
arr2 = np.arange(1,100000001)

start = time.time()
arr3 = arr1 + arr2 
end = time.time()

print(f"First 5 element of arr3 = {arr3[:5]}")
print(f"Last 5 element of arr3 = {arr3[-1:-6:-1]}")
print(f"Time Take to complete operation of 2 arrays addition is {end-start} sec")
```
